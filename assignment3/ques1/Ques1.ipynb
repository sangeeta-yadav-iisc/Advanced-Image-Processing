```python
import numpy as np 
from scipy.integrate import quad
from math import exp, pi
import matplotlib.pyplot as plt
### MMSE estimation for laplacian source

Vx = 1
Vz = 0.1
lim = 4*Vx

y = np.linspace(-lim,lim,100)
```

```python
sols = []
for yi in y:
    num = lambda x: x*exp(-np.abs(x)/Vx)*np.exp((-0.5/Vz)*((yi-x)**2))
    den = lambda x: exp(-np.abs(x)/Vx)*np.exp((-0.5/Vz)*((yi-x)**2))
    nums,_ =quad(num,-abs(yi)-0.1,abs(yi)+0.1)
    dens,_ =quad(den,-abs(yi)-0.1,abs(yi)+0.1)  
    sols.append(nums/dens)                                                                                 

fig,ax = plt.subplots(figsize=(6,6))
ax.set_xlabel('y',fontsize=20)
ax.set_ylabel('$\hat{x}$',fontsize=20)
ax.plot(y,sols,label = "Denoised Image")
ax.plot(y,y,'--',label = "Identity Function")
ax.legend()
```
